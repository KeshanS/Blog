```python
'''In this example, I use the data on calls for service from New Orleans.
This data are available here: https://data.nola.gov/
I use the dataset from 2015. I downloaded the CSV files.
'''
```

```python
# I import pandas.
import pandas as pd
```

```python
# I read the CSV file.
twenty15_df = pd.read_csv("Calls_for_Service_2015.csv")
twenty15_df.head()
```

```python
'''Checking if any of your records are duplicates is very easy.
Use:
df.duplicated()'''
```

```python
twenty15_df.duplicated()
# I can see that here I have no duplicate records as all of the values are displayed as False.
```

```python
'''You can also check just for duplicates in a particular column.
df.duplicated('col1')
This returns a Boolean.'''
```

```python
twenty15_df.duplicated('TypeText')
```

```python
'''If I want to get rid of duplicates, I can drop them.
df.drop_duplicates()'''
```

```python
twenty15_df.drop_duplicates()
```

```python
'''Similarly to checking for duplicates, I can also drop duplicates from a particular column.'''
```

```python
twenty15_df.drop_duplicates(['TypeText'])
```

```python
'''In additiona to dropping duplicates from the column I want, I also get to choose if I keep the first record,
the last or if I want to drop all duplicates altogether. This can be done by adding keep and choosing it as first,
last or False. False drops all the duplicates, first keeps the first one and last keeps the last record.
The default is first.'''
```

```python
twenty15_df.drop_duplicates(['TypeText'], keep='first')
```

```python
twenty15_df.drop_duplicates(['TypeText'], keep='last')
```

```python
twenty15_df.drop_duplicates(['TypeText'], keep=False)
```
