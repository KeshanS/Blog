When doing data science in Python, you may be asked to analyse the data that's in CSV or Excel file.
You need to be able to read this file into Python. Using pandas will help you to automatically convert it into a DataFrame.

Firstly, I am going to show you how to both import and export CSV files.

dframe = pd.read_csv("file_name.csv")

```python
# Importing relevant libraries.
import pandas as pd
```

```python
# Importing a CSV file and viewing the first 5 rows.
dframe = pd.read_csv("Calls_for_Service_2015.csv")
dframe.head()
```

You can also specify the delimiter - not always will it be a comma (as in this file).
dframe = pd.read_table('file_name.csv',sep=',')

```python
# Importing and viewing the first 5 rows again.
dframe = pd.read_table('Calls_for_Service_2015.csv', sep=',')
dframe.head()
```

If you have a table without column names, you can specify that you don't want to use the first row as the header by passing the argument header=None

df = pd.read_csv("file_name.csv", header=None)

```python
# Example

dframe = pd.read_csv('Calls_for_Service_2015.csv',header=None)
dframe.head()

# You can see that here I get a warning - this is because the first row contains column names (which are strings)
# whereas MapX and MapY values are actually integers.
```

It is also possible to specify the number of rows to be read - for example, I only want the first 10 rows:

pd.read_csv('file_name.csv',nrows=10)

```python
pd.read_csv('Calls_for_Service_2015.csv',nrows=10)
```

It is also really easy to export the DataFrame as a CSV file - for example, after you've performed some operations on it.

dframe.to_csv("file_name.csv")

```python
# Exporting the DataFrame.
dframe.to_csv("dframe.csv")
```

You can also specify what delimiter you want to use by adding the argument sep='delimiter'

```python
# Here I decided to use a question mark for a delimiter.
dframe.to_csv("dframe.csv", sep='?')
```
