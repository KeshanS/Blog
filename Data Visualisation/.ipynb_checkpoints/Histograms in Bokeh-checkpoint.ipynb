```python
# Importing the usual modules.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from bokeh.plotting import *
from bokeh.charts import Histogram, defaults, vplot, hplot, output_file, show
output_notebook()
%matplotlib inline
```

```python
# Importing the dataset
houseprices = pd.read_csv('kc_house_data.csv')
houseprices.head()
```

Creating a histogram in Bokeh:  

p = Histogram(DataFrame, 'price', values='price')

```python
p = Histogram(houseprices['price'], bins=10)

show(p)
```

# Labelling the axis and adding the title to the plot.

Pass the arguments:  
xlabel = 'Label for x axis'  
ylabel = 'Label fo y axis'

```python
p = Histogram(houseprices['price'], bins=10, xlabel='House price', ylabel='Frequency')

show(p)
```

# Changing the plot width and height, and bar width

Pass the arguments:  
plot_width  
plot_height  
bar_width

```python
p = Histogram(houseprices['price'], bins=10, plot_width = 450, plot_height = 450)

show(p)
```

```python
p = Histogram(houseprices['price'], bins=10, bar_width = 0.5)

show(p)
```
