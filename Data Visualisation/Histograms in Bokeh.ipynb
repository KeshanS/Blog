```python
# Importing the usual modules.
import pandas as pd
import numpy as np
from bokeh.plotting import *
from bokeh.charts import Histogram, defaults, vplot, hplot, output_file, show
output_notebook()
%matplotlib inline
```

```python
# Importing the dataset
houseprices = pd.read_csv('kc_house_data.csv')
houseprices.head()
```

### Creating a histogram in Bokeh:  

p = Histogram(DataFrame, values='column_name')<br>
show(p)

```python
p = Histogram(houseprices, values='price')

show(p)
```

### Changing the number of bins

p = Histogram(DataFrame, values='column_name', bins=number)<br>
show(p)

```python
p = Histogram(houseprices, values='price', bins=10)

show(p)
```

### Adding title to the plot

There are two ways to add the title to the plot:<br>
1) Pass the argument 'title' within the function: <br>
2) Add p.title.text = 'your title' underneath <br>

```python
# The first way of doing this:

p = Histogram(houseprices, values='price', bins=10, title = 'House prices')

show(p)
```

```python
# The first way of doing this:

p = Histogram(houseprices, values='price', bins=10)
p.title.text = 'House prices'

show(p)
```

# Labelling the axis and adding the title to the plot.

Pass the arguments:  
xlabel = 'Label for x axis'  
ylabel = 'Label fo y axis'

```python
p = Histogram(houseprices['price'], bins=10, xlabel='House price', ylabel='Frequency')

show(p)
```

### Changing the range on the x and y axis.
Pass the following arguments: <br>
p.x_range.start = 0 <br>
p.x_range.end = 1000<br><br>
**Similarly, for the y axis:** <br>
p.y_range.start = 0 <br>
p.y_range.end = 1000 <br>

```python
p = Histogram(houseprices['price'], bins=10, xlabel='House price', ylabel='Frequency')
p.x_range.start = 0
p.x_range.end = 100000
p.y_range.start = 0
p.y_range.end = 1000

show(p)
```

### Changing the plot width and height

Pass the arguments:  
plot_width  
plot_height  

```python
p = Histogram(houseprices['price'], bins=10, plot_width = 450, plot_height = 450)

show(p)
```

### Grouping by colour

```python
p = Histogram(houseprices['price'], bins=10, plot_width = 450, plot_height = 450, legend=True)

show(p)
```
