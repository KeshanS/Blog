```python
# The usual imports
import pandas as pd
```

```python
'''I read a table that I have from a different exercise. It contains names of the days of the week of departure,
days of the week of return and flight price.
'''
dframe = pd.read_csv("grouped3df.csv")

# Pivoting this table first.

pivot_table = dframe.pivot('day_of_week_dep', 'day_of_week_ret', 'total')
pivot_table
```

The order of columns obviously does not make sense. Neither does the order of rows.
Re-ordering columns is easy:
dframe = dframe[['col1', 'col2', 'col3']]
Remember to include all of the column names - otherwise this function will not work

```python
# Changing the order of columns.
pivot_table = pivot_table[['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']]
pivot_table
```

Doing the same for indices is really easy - you do it as follows:

dframe = dframe.reindex[['index1', 'index2', 'index3']]

```python
# Re-indexing the rows.

pivot_table = pivot_table.reindex(['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun'])
pivot_table
```
