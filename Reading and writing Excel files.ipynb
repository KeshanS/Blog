In this post, a Kaggle dataset on 2016 US Elections was used (https://www.kaggle.com/benhamner/d/benhamner/2016-us-election/primary-results-sample-data/output).
This dataset has been converted from a CSV file to an Excel file and two sheets have been added with votes for Hilary Clinton (HilaryClinton) and Donald Trump (DonaldTrump). The first sheet (All) contains the original dataset.

```python
# Usual imports
import pandas as pd
```

Reading an Excel file is very similar to reading a CSV file:

dframe = pd.read_excel("file_name.xlsx")

By default, the first sheet of the file will be imported.

```python
# Reading an Excel file and viewing the first 5 rows.
results = pd.read_excel("2016_results.xlsx")
results.head()
```

You can easily choose a specific sheet from the file to be imported. If you know the sheet name, it is done as follows:

dframe = pd.read_excel("file_name.xlsx", sheetname="Sheet_name")

```python
# Importing the sheet with results for Donald Trump only.
resultsdt = pd.read_excel("2016_results.xlsx", sheetname="DonaldTrump")
resultsdt.head()
```

If you are unsure of what the sheet name is but you know how many sheets you have, you can pass a number of the sheet instead of a sheet name. Please note that the sheets start from 0 (similar to indices in pandas), not from 1.

dframe = pd.read_excel("file_name.xlsx", sheetname=number)

```python
# Viewing the second sheet of the Excel file.
results2 = pd.read_excel("2016_results.xlsx", sheetname=1)
results2.head()
```

Sometimes, the first row does not contain the column names. In this instance, you pass an argument of header=None:
dframe = pd.read_excel("file_name.xlsx", header=None)

```python
# Reading the file and not using the first row as column names.
results3 = pd.read_excel("2016_results.xlsx", header=None)
results3.head()
```

By passing the header argument, you can use the nth row as the column names.
dframe = pd.read_excel("file_name.xlsx", header=n)

```python
# Reading the file and using the third column as the column names.
results4 = pd.read_excel("2016_results.xlsx", header=2)
results4.head()
```

You can use different columns for the row labels by passing the index_col argument as number.
dframe = pd.read_excel("file_name.xlsx", index_col=number)

```python
# Reading the data and using the row index as a county i.e. the 3rd column.
results5 = pd.read_excel("2016_results.xlsx",  index_col=2)
results5.head()
```

Sometimes, you may want to exclude some of the rows. If you want to skip n number of first rows, just pass the argument of skiprows:

dframe = pd.read_excel("file_name.xlsx", skiprows=n)

```python
# Reading the file and skipping the first two rows.
results7 = pd.read_excel("2016_results.xlsx",  skiprows=2)
results7.head()
```

Writing an Excel file is very easy:
dframe.to_excel('file_name.xlsx')

```python
# Writing an Excel file
results.to_excel('results.xlsx')
```

If you don't want to include the index name (for example, here it is a number so it may be meaningless for future use/analysis), you can just pass another argument, setting index as False:
dframe.to_excel('file_name.xlsx', index=False)

```python
# Writing an Excel file without the index names
results.to_excel('results2.xlsx', index=False)
```
